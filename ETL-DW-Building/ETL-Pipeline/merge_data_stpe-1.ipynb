# Bronze Layer - Merge Raw data -  Google Patent & OPS 

```python
import json
```

```python
with open('raw-data/google_patent/google_patent_raw-data.json','r') as f:
    data = json.load(f)
```

```python
for patent in data :
    patent.pop('country')
```

```python
for patent in data:
    patent['application date']=""
```

```python
data[0]
```

```python
def process_name(name_with_country):
    return name_with_country.split("[")[0].strip().title()
```

```python
with open('raw-data/OPS/data1.json','r') as f:
    data1 = json.load(f)

with open('raw-data/OPS/data2.json','r') as f:
    data2 = json.load(f)

with open('raw-data/OPS/data3.json','r') as f:
    data3 = json.load(f)

with open('raw-data/OPS/data4.json','r') as f:
    data4 = json.load(f)
```

```python
ops = data1+ data2 + data3+ data4
```

```python
for patent in ops:
    inventors=[]
    for inventor in patent['inventor_name']:
        inventors.append(process_name(inventor))
    
    applicants=[]
    for applicant in patent['applicants']:
        applicants.append(process_name(applicant))

    patent['title'] = patent['title'].strip().capitalize()
    
    patent['inventor_name'] = inventors
    patent['applicants'] = applicants
```

```python
with open('raw-data/OPS/ops_patents.json','w') as f:
    json.dump(ops,f,indent=4)
```

```python
patents = ops+data
```

```python
with open('raw-data/patents.json','w') as f:
    json.dump(patents,f,indent=4)
```
